# Scatterplot

```python
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex',
    style='smoker',
    size='size'
)
```

```python
sns.barplot(
    data= titanic,
    x='Pclass',
    y='Age',
    hue='Sex'
)
```

```python
sns.boxplot(
    data= titanic,
    x='Sex',
    y='Age',
    hue='Survived'
)
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'], hist = False)
sns.distplot(titanic[titanic['Survived']==1]['Age'])
```

```python
sns.heatmap( pd.crosstab(
    titanic['Survived'],
    titanic['Pclass']))
```

```python
sns.heatmap(
    tips.pivot_table(values= 'tip', index='smoker', columns='size')
)
```

```python
sns.lineplot(
    x = flights['month'], y = flights['passengers'], marker='o')
```
